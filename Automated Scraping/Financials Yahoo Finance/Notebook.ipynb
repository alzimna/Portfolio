```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from urllib.error import HTTPError
```

```python
a = pd.read_excel("https://github.com/alzimna/Portfolio/raw/refs/heads/main/Automated%20Scraping/Financials%20Yahoo%20Finance/nama.xlsx")
a = a[["Kode"]]
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
nama = a["Kode"]
```

```python
symbol = nama[0]+".JK"
url = "https://finance.yahoo.com/quote/"+symbol+"/financials?p="+symbol
response = requests.get(url,headers = headers)
doc = BeautifulSoup(response.text, "html.parser")
columnname = doc.find_all(class_="row yf-1yyu1pc")[0].text.split()[1:]
d = dict()
for item in columnname :
    d[item] = []
for i in range(len(nama)) :
    symbol = nama[i]+".JK"
    url = "https://finance.yahoo.com/quote/"+symbol+"/financials?p="+symbol
    response = requests.get(url,headers = headers)
    doc = BeautifulSoup(response.text, "html.parser")
    temp = doc.find_all(class_="row lv-0 yf-t22klz")
    j = 0
    while j>=0 :
        if temp[j].find_all(class_="rowTitle yf-t22klz")[0].string == "Basic EPS" :
            break
        j+=1
    temp1 = temp[j].text.split()[2:]
    for k in range(len(columnname)) :
        d[columnname[k]].append(temp1[k])
```

```python
d["Kode"] = nama
df = pd.DataFrame(d)
df.set_index("Kode")
```
