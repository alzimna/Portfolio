```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from urllib.error import HTTPError
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
nama = a["Kode"]
```

```python
EPS_2024 = []
EPS_2023 = []
for i in range(len(nama)) :
    symbol = nama[i]+".JK"
    url = "https://finance.yahoo.com/quote/"+symbol+"/financials?p="+symbol
    response = requests.get(url,headers = headers)
    doc = BeautifulSoup(response.text, "html.parser")
    temp = doc.find_all(class_="row lv-0 yf-t22klz")
    j = 0
    while j>=0 :
        if temp[j].find_all(class_="rowTitle yf-t22klz")[0].string == "Basic EPS" :
            break
        j+=1
    temp1 = temp[j].text.split()
    if isinstance(temp1[3], float) :
        temp2024 = float(temp1[3])*1000
    else :
        temp2024 = temp1[3]

    if isinstance(temp1[4], float) :
        temp2023 = float(temp1[4])*1000
    else :
        temp2023 = temp1[4]

    EPS_2024.append(temp2024)
    EPS_2023.append(temp2023)
```

```python
d = {"Kode" : nama, "EPS 2024" : EPS_2024,"EPS 2023" : EPS_2023}
df = pd.DataFrame(d)
```
